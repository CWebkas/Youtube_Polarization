## Classify the Youtube Videos
After obtaining the cideous and their recommendations, it might be interesting, if there is a difference in their degree of recommendations. I had found, that number of views as well as number of likes don't influence the number of recommendations. Therefore, the content might have some relevance in terms of recommendations.
First, I install the required packages, then I classify into three clusters, based on similarity in tags. Close distance indicates similar tags. The closer two units, the more likely they appear within a cluster. Based on that, I create a dendrogram. Three clusters show the clearest separations, which is also obvious when looking at the following workclouds. Two clusters combine too much differing content. (i.e. H.C. Strache with Tagesschau). Four are inefficient, due to unnecessary separation.

```python
import pandas as pd
import json
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.decomposition import NMF 
from sklearn.preprocessing import normalize
from pprint import pprint
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
```

```python
from scipy.cluster.hierarchy import linkage, dendrogram
```

```python
from scipy import cluster
```

```python
import numpy as np
```

```python
from itertools import chain
```

```python
import seaborn as sns
```

```python
df_frame = pd.read_pickle('Data/Recommendations_pd')
```

```python
with open('Data/nested_recommendations', 'r', encoding = 'utf-8') as f:
    nested2 = json.load(f)
```

```python
# Get a List of Tags out of my clean list of dicts for the Learning Model:
taglist = [i['Tags'] for i in nested2]


# Final list to append tags to
unnested_taglist = []

# Get every tag out of the taglist created above
for i in taglist:
    for elem in i:
        # Append to final list
        unnested_taglist.append(elem)
```

```python
textlst = []
for i in taglist:
    text = " ".join(unit for unit in i)
    textlst.append(text)
```

```python
# Create the vectorizer method
tfidf_vec = TfidfVectorizer()

# Transform the text into tf-idf vectors
text_tfidf = tfidf_vec.fit_transform(textlst)
```

```python
print(text_tfidf)
```

```python
# Import NMF


# Create an NMF instance: model
model = NMF(n_components=3)

# Fit the model to articles
model.fit(text_tfidf)

# Transform the articles: nmf_features
nmf_features = model.transform(text_tfidf)

# Print the NMF features
print(nmf_features)
```

```python
# Import NMF


# Create an NMF instance: model
model = NMF(n_components=20)

# Fit the model to articles
model.fit(text_tfidf)

# Transform the articles: nmf_features
nmf_features = model.transform(text_tfidf)

# Print the NMF features
print(nmf_features)
```

```python
ids = df_frame['Video Id']
```

```python
df = pd.DataFrame(nmf_features, index = ids)
```

```python
print(df.loc['2gFQ6qhWea8'])
```

```python
norm_features = normalize(nmf_features)
```

```python
df = pd.DataFrame(norm_features, index = ids)
```

### Test the similarities towards a specific video

```python
article = df.loc['2gFQ6qhWea8']
```

```python
similarities = df.dot(article)
```

### Print the most similar videos and their cosine similarity

```python
print(similarities.nlargest())
```

```python
pprint(similarities['ZtKFu9HzFsc'])
```

```python
mergings = linkage(norm_features, 'complete')
```

### Display the cluster structure, cutting tree at level 1.3

```python
figure(figsize=(20,20))
dendrogram(mergings, leaf_rotation = 90)

plt.show()
```

```python
cutree = cluster.hierarchy.cut_tree(mergings, height = 1.3).tolist()
```

```python
total = []
for i in cutree:
    for elem in i:
        total.append(elem)
```

```python
cluster_df = pd.DataFrame(total)
```

```python
cluster_df
```

```python
cluster_df = cluster_df.rename(columns={0:'Cluster'})
```

### Appending cluster to dataframe

```python
final_frame = pd.concat([df_frame, cluster_df], axis = 1, join = 'inner')
```

```python
final_frame.reset_index()
```

### Display the clusters
I am concerned, by what tags the clusters are grouped. Therefore, I create 3 different dataframes, depending on their cluster membership.

```python
zero_clust = final_frame.loc[final_frame['Cluster'] == 0]
```

```python
one_clust = final_frame.loc[final_frame['Cluster'] == 1]
```

```python
two_clust = final_frame.loc[final_frame['Cluster'] == 2]
```

### get the tags for every Cluster
Summarizing the tags lets me put them into a fluent text, enabling the production of a word cloud.

```python
zero_dict = zero_clust['Tags']
one_dict = one_clust['Tags']
two_dict = two_clust['Tags']
#three_dict = three_clust['Tags']
```

```python
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
```

```python
zero_list = zero_dict.tolist()
one_list = one_dict.tolist()
two_list = two_dict.tolist()
#three_list = three_dict.tolist()
```

### Creating fluent Texts
In order to produce wordcluds, I require fluent text.

```python
unnested_zero = []
for i in zero_list:
    for elem in i:
        # Append to final list
        unnested_zero.append(elem)
text0 = " ".join(unit for unit in unnested_zero)
```

```python
unnested_one = []
for i in one_list:
    for elem in i:
        # Append to final list
        unnested_one.append(elem)
text1 = " ".join(unit for unit in unnested_one)
```

```python
unnested_two = []
for i in two_list:
    for elem in i:
        # Append to final list
        unnested_two.append(elem)
text2 = " ".join(unit for unit in unnested_two)
```
